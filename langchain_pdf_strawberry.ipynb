```python
# 데이터 불러와서 페이지마다 분할
loader = PyPDFLoader("/content/drive/MyDrive/strawberry_test.pdf")
pages = loader.load_and_split()

# 허깅페이스 한국어 fine트레이닝된 임베딩 모델 불러옴
model_name = "jhgan/ko-sbert-nli"
model_kwargs = {'device': 'cpu'}
encode_kwargs = {'normalize_embeddings': True}
hf = HuggingFaceEmbeddings(
    model_name=model_name,
    model_kwargs=model_kwargs,
    encode_kwargs=encode_kwargs
)

# 텍스트를 분할하여 벡터db에 임베딩해줌
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=200)
splits = text_splitter.split_documents(pages)
vectorstore = Chroma.from_documents(documents=splits, embedding=hf, persist_directory="./chroma_db")

# 사용 llm으로는 gpt 3.5 사용
retriever = vectorstore.as_retriever()
llm = ChatOpenAI(model_name="gpt-3.5-turbo", temperature=0)

# 랭체인 LCEL를 사용하여 chain
from langchain.chains import RetrievalQA
qa_chain = RetrievalQA.from_chain_type(
    llm,
    retriever=vectorstore.as_retriever()
)
question = "딸기에 대형유통업체에 대해 자세하게 설명해줘"
result = qa_chain({"query": question})
result["result"]
```

```python
from langchain.chains import RetrievalQA
qa_chain = RetrievalQA.from_llm(
    llm,
    retriever=vectorstore.as_retriever()
)
question = "딸기 유통비용명세표"
result = qa_chain({"query": question})
result["result"]
```

```python
from langchain.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
output_parser = StrOutputParser()
prompt = ChatPromptTemplate.from_messages(
    [
        (
            "system",
            """
            you are a strawberry expert, you need to help the questions : {context}
            """,
        ),
        ("human", "{question}"),
    ]
)

chain = (
    {
        "context": retriever,
        "question": RunnablePassthrough(),
    }
    | prompt
    | llm
    #| output_parser

)
result = chain.invoke('딸기 가격 추이')
print(result)
```

```python
for s in chain.stream("딸기 가격 추이"):
    print(s.content, end="", flush=True)
```
